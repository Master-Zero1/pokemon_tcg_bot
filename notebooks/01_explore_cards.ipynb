```python
#List of 50 cards
for card in all_cards[:50]:
    print(vars(card))
```

```python
#List of 5 cards from mid
for card in all_cards[34:39]:
    print(vars(card))
```

```python
#First pokemon card
for card in all_cards:
    if card.hp>0:
        print(vars(card))
        break
```

```python
#info about attacks

#Total attacks
print(f"Total attacks: {len(all_attacks)}")

#Attacks for the first pokemon card
attack_table = {a.attackId: a for a in all_attacks}
scrafty = all_cards[20] #(index is 20)
for attack_id in scrafty.attacks:
    print(vars(attack_table[attack_id]))
```

```python
#First trainer card
for card in all_cards:
    if card.cardType==3:
        print(vars(card))
        break
```

```python
print(f"Total attacks: {len(all_attacks)}")
print("\n--- First attack ---")
print(vars(all_attacks[0]))

print("\n--- Attacks for Scrafty (cardId 21) ---")
attack_table = {a.attackId: a for a in all_attacks}
scrafty = all_cards[20]  # cardId 21 is index 20
for attack_id in scrafty.attacks:
    print(vars(attack_table[attack_id]))
```
